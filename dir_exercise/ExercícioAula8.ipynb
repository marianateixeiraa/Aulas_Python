1)Resolução numérica de equações diferenciais. Imagine que tem um/a maratonista treinando, que num trecho de $5 \mathrm{km}$ tem velocidade quase constante de $12 \mathrm{km/h}$. Produza uma lista com a distância percorrida em intervalos de 1 minuto.

```python
import matplotlib.pyplot as plt

def distancia_mru(t, ti, v, di):
    di = v*(t - ti) + di
    return di
    
t = 0  # min
lista_t = [t] 
while t < 25: 
    t = t + 1
    lista_t.append(t) 

v = 200  # m/min
lista_di = []
for t in lista_t:
    di = distancia_mru(t, 0, v, 0)
    lista_di.append(di/(1000)) # km

print(lista_di)
print(lista_t)

plt.plot(lista_t,lista_di)
plt.xlabel("tempo (minutos)")
plt.ylabel("distância ((km))")
plt.show() 
```

3) Considere um péndulo simples num plano. O comprimento da linha é $1$ metro, a massa é $0,2 \mathrm{kg}$ e a aceleração de gravidade $9,8 \mathrm{m}/\mathrm{s}^2$. Faça um gráfico ou crie uma tabela para a posição da massa em coordenadas cartesianas em função do tempo, utilizando listas.

```python
import matplotlib.pyplot as plt

def pendulum_angle(amplitude, length, time, g):
    
    import math
    angle = amplitude*math.cos(math.sqrt(g/length)*time)
    return angle 

def pendulum_period(length, g):
    
    import math
    period = 2*math.pi*math.sqrt(length/g)
    return period

def pendulum_x(amplitude, length, time, g):
    
    import math
    a = pendulum_angle(amplitude, length, time, g)
    x = length*math.sin(a)
    return x 

def pendulum_y(amplitude, length, time, g):
    
    import math
    a = pendulum_angle(amplitude, length, time, g)
    y = - length*math.cos(a)
    return y


time = 0
list_time = []
list_x = []
list_y = []
while time <= pendulum_period(1, 9.8):
    time = time + 0.01
    list_time.append(time)
    a = pendulum_x(0.4, 1, time, 9.8)
    b = pendulum_y(0.4, 1, time, 9.8)
    list_x.append(a)
    list_y.append(b)


print(list_time)
print(list_x)
print(list_y)

plt.plot(list_time, list_x)
plt.show()

plt.plot(list_time, list_y)
plt.show()
```
