```python
%%writefile Projeto.py

import matplotlib.pyplot as plt
import math
import numpy as np



def eq_EDO(t,C,k) :                      
  """
  Recebe um valor t referente ao tempo, um valor C referente à temperatura em graus Celsius e um valor k referente à regulamentação.
  
  Retorna a derivada da temperatura em relação ao tempo.
  """
  return (1/10)*(C-20)*(22-C)*(C-26) - k


def Euler(t,C0,k):
  """
  Recebe uma lista t cujos elementos são os valores de um intervalo referente ao tempo decorrido, um valor C0 referente à temperatura inicial em graus Celsius e um valor k referente à regulamentação.
  
  Retorna uma lista de valores da temperatura em função do tempo de acordo com a função eq_EDO.
  """

  C=[C0]    
  dt = t[1]-t[0] # como os intervalos são regulares, basta identificar a diferênça entre o primeiro par.                                          

  for i in range(1,len(t)) :                             
    Ci = C[i-1]  + eq_EDO(t[i-1],C[i-1],k) *dt               
    C.append(Ci)

  return C


def temp(C0,k) :
  """
  Recebe um valor C0 referente à temperatura inicial em graus Celsius e um valor k referente à regulamentação.
  
  Plota um gráfico Temperetura x Tempo para os respectivos valores de k e C0 num intervalo de tempo de 50 anos.
  """
  t = np.arange(0,50,0.1)
  C = Euler(t, C0,round(k,2))
  plt.plot(t,C,label=f'k = {round(k,2)}')

  # legendas do gráfico
  plt.xlabel('tempo [anos]')
  plt.ylabel('temperatura °C')
  plt.legend()   
```

```python
import Projeto
from matplotlib import cm

def dicionario_Euler(C0_inicial,C0_final,dC,k_inicial,k_final,dk): 
  """
  Recebe um valor C0_inicial referente ao primeiro valor tomado para a temperatura inicial (em graus Celsius), um valor C0_final referente ao último valor 
  tomado para a temperatura inicial (em graus Celsius) e um valor dC referente ao espaçamento entre os valores tomados para a temperatura inicial. 
  Recebe também um valor k_inicial referente ao primeiro valor tomado para a regulamentação, um valor k_final referente ao último valor tomado para a regulamentação
  e um valor dk referente ao espaçamento entre os valores tomados para a regulamentação.

  Retorna um dicionário cujas chaves são os valores de C0 referentes ao intervalo dado e os valores correspondentes a essas chaves são dicionários.
  Estes dicionários são aqueles cujas chaves são os valores de k referentes ao intervalo dado e os valores correspondentes a essas chaves são listas obtidas através da função Euler.
  """
  t = np.arange(0,50,0.1)
  dic = {}
  for Ci in np.arange(C0_inicial,C0_final,dC):
    dic[Ci] = {}
    for ki in np.arange(k_inicial,k_final,dk) :
      dic[Ci][round(ki,2)] = Projeto.Euler(t,Ci,ki)
  return dic

def grafico_temperatura_tempo(C0_inicial,C0_final,dC,k_inicial,k_final,dk):
  """
  Recebe um valor C0_inicial referente ao primeiro valor tomado para a temperatura inicial (em graus Celsius), um valor C0_final referente ao último valor 
  tomado para a temperatura inicial (em graus Celsius) e um valor dC referente ao espaçamento entre os valores tomados para a temperatura inicial. 
  Recebe também um valor k_inicial referente ao primeiro valor tomado para a regulamentação, um valor k_final referente ao último valor tomado para a regulamentação
  e um valor dk referente ao espaçamento entre os valores tomados para a regulamentação.

  Plota o gráfico Temperetura x Tempo para os respectivos valores de k e C0 da rede dada como entrada num intervalo de tempo de 50 anos.
  """

  for Ci in np.arange(C0_inicial,C0_final,dC):
    for ki in np.arange(k_inicial,k_final,dk):
      Projeto.temp(Ci,ki)

def Temperatura_equilibrio(C0,k) :
  """
  Recebe um valor C0 referente à temperatura inicial em graus Celsius e um valor k referente à regulamentação.
  
  Retorna um valor correspondente à temperatura de equilíbrio atingida num intervalo de 50 anos a partir dos valores de temparatura inicial e regulamentação dados.
  """
  t = np.arange(0,50,0.1)
  solucao_eq_dif = Projeto.Euler(t,C0,k)
  
  for i in range(len(solucao_eq_dif)):
    if abs(solucao_eq_dif[-1] - solucao_eq_dif[i]) <= 0.01 : # como sabemos que o platô é atingido no ano 50, utilizamos o último valor da lista de soluções para a comparação.
     
      return solucao_eq_dif[i]
          

def Tempo_Temperatura_equilibrio(C0,k) :
  """
  Recebe um valor C0 referente à temperatura inicial em graus Celsius e um valor k referente à regulamentação.
  
  Retorna um valor correspondente ao momento (em anos) no qual a temperatura de equilíbrio é atingida num intervalo de 50 anos a partir dos valores de temparatura inicial e regulamentação dados.
  """
  t = np.arange(0,50,0.1)
  T_eq = Projeto.Euler(t,C0,k).index(Temperatura_equilibrio(C0,k))
    
  return T_eq*0.1


def dicionario_C0_k_Teq(C0_inicial,C0_final,dC,k_inicial,k_final,dk) :
  """
  Recebe um valor C0_inicial referente ao primeiro valor tomado para a temperatura inicial (em graus Celsius), um valor C0_final referente ao último valor 
  tomado para a temperatura inicial (em graus Celsius) e um valor dC referente ao espaçamento entre os valores tomados para a temperatura inicial. 
  Recebe também um valor k_inicial referente ao primeiro valor tomado para a regulamentação, um valor k_final referente ao último valor tomado para a regulamentação
  e um valor dk referente ao espaçamento entre os valores tomados para a regulamentação.

  Retorna um dicionário cujas chaves são os valores de C0 referentes ao intervalo dado e os valores correspondentes a essas chaves são dicionários.
  Estes dicionários são aqueles cujas chaves são os valores de k referentes ao intervalo dado e os valores correspondentes a essas chaves são as temperaturas de equilibrio respectivas.
  """
  dic = {}
  for Ci in np.arange(C0_inicial,C0_final,dC):
    dic[Ci] = {}
    for ki in np.arange(k_inicial,k_final,dk) :
      dic[Ci][round(ki,2)] =Temperatura_equilibrio(Ci,ki)
  return dic

def grafico_Teq_x_k(C0_inicial,C0_final,dC,k_inicial,k_final,dk) :
  """ 
  Recebe um valor C0_inicial referente ao primeiro valor tomado para a temperatura inicial (em graus Celsius), um valor C0_final referente ao último valor 
  tomado para a temperatura inicial (em graus Celsius) e um valor dC referente ao espaçamento entre os valores tomados para a temperatura inicial. 
  Recebe também um valor k_inicial referente ao primeiro valor tomado para a regulamentação, um valor k_final referente ao último valor tomado para a regulamentação
  e um valor dk referente ao espaçamento entre os valores tomados para a regulamentação.
  
  Plota o grafico Temperatura de equilibrio x regulamentação para valores em rede dados como entrada. 

  obs: em alguns casos os graficos podem estar sobrepostos
  """
  dic = dicionario_C0_k_Teq(C0_inicial,C0_final,dC,k_inicial,k_final,dk)
  Ci = list(dic.keys())
  for j in Ci :
    ki = []
    T_eq = []
    for chave in dic[j].keys() :
      ki.append(chave)  
    for valor in dic[j].values() :
      T_eq.append(valor)
    plt.plot(ki,T_eq,label = f'C0 = {j}')


  # legendas do gráfico
  plt.xlabel('regulamentação')
  plt.ylabel('temperatura °C')
  plt.legend()      

def grafico_dispersão(C0_inicial,C0_final,dC,k_inicial,k_final,dk) :
  """
  Recebe um valor C0_inicial referente ao primeiro valor tomado para a temperatura inicial (em graus Celsius), um valor C0_final referente ao último valor 
  tomado para a temperatura inicial (em graus Celsius) e um valor dC referente ao espaçamento entre os valores tomados para a temperatura inicial. 
  Recebe também um valor k_inicial referente ao primeiro valor tomado para a regulamentação, um valor k_final referente ao último valor tomado para a regulamentação
  e um valor dk referente ao espaçamento entre os valores tomados para a regulamentação.

  Plota o grafico de dispersão dos valores de temperature de equilibrio x regulamentação para cada valor de temperatura inicial 

  obs:muitos pontos estarão sobrepostos por isso fizemos um certo degrade para podermos identificar um pouco melhor
  """
  from matplotlib import cm
  dic = dicionario_C0_k_Teq(C0_inicial,C0_final,dC,k_inicial,k_final,dk)
 
  colors = cm.rainbow(np.linspace(0, 1, len(dic))) 
  plt.figure(figsize=(15,9))            # formatação para cor
  for i, dic_values in enumerate(dic.values()):               # enumera os valores respectivo para cada chave de C0
    k = dic_values.keys()
    T_eq = dic_values.values()
    plt.scatter(k,T_eq,color=colors[i],s= (C0_final*3/dC) - ((C0_final - C0_inicial)/dC)*i)        # S é degrade do grafico 

  
  plt.legend(dic.keys())                                      # as chaves do dicionario principal para suas respectivas cores de acordo com a enumeração
  plt.xlabel('K')
  plt.ylabel('temperatura Equilibrio (°C)')
  plt.show()

```
